```python
# List of numeric columns to check for outliers
numeric_columns = ['person_age','person_income', 'loan_amnt', 'credit_score', 'loan_int_rate', 'loan_percent_income']

# Create a copy of the dataset to avoid modifying the original
df_cleaned = df.copy()

# Loop through each numeric column
for column in numeric_columns:
    # Calculate Q1, Q3, and IQR
    Q1 = df_cleaned[column].quantile(0.25)
    Q3 = df_cleaned[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter data to remove outliers
    df_cleaned = df_cleaned[(df_cleaned[column] >= lower_bound) & (df_cleaned[column] <= upper_bound)]

# Print the shape of the dataset before and after cleaning
print("Original dataset shape:", df.shape)
print("Cleaned dataset shape:", df_cleaned.shape)
```

```python
# Re-plot box plots after removing outliers
plt.figure(figsize=(15, 5))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(1, len(numeric_columns), i)
    sns.boxplot(data=df_cleaned, y=column, color='salmon')
    plt.title(f'Box Plot of {column}')
    plt.ylabel(column)
plt.tight_layout()
plt.show()
```

```python
# Numeric columns to include in correlation analysis
numeric_columns = ['person_age', 'person_income', 'credit_score', 'loan_amnt', 'loan_int_rate', 'loan_percent_income', 'loan_status']

# Compute the correlation matrix
correlation_matrix = df_cleaned[numeric_columns].corr()

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix', fontsize=16, fontweight='bold', pad=15)
plt.show()
```

```python
from scipy.stats import chi2_contingency

# List of categorical columns
categorical_columns = ['person_gender', 'person_education', 'person_home_ownership', 'loan_intent', 'previous_loan_defaults_on_file']

# Loop through each categorical column
for column in categorical_columns:
    # Create a contingency table
    contingency_table = pd.crosstab(df_cleaned[column], df_cleaned['loan_status'])

    # Perform Chi-Square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Print the results
    print(f"Chi-Square Test for {column}:")
    print(f"Chi2 Statistic = {chi2:.2f}, p-value = {p:.4f}")
    if p < 0.05:
        print(f"  -> Significant association with loan_status (p < 0.05)\n")
    else:
        print(f"  -> No significant association with loan_status (p â‰¥ 0.05)\n")
```

```python
import numpy as np
from scipy.stats import chi2_contingency

def cramers_v(confusion_matrix):
    # Ensure the input is a valid NumPy array
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.to_numpy().sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))

# Loop through each categorical column
for column in categorical_columns:
    # Create a contingency table
    contingency_table = pd.crosstab(df_cleaned[column], df_cleaned['loan_status'])

    # Calculate CramÃ©r's V
    cramers_value = cramers_v(contingency_table)

    # Print the result
    print(f"CramÃ©r's V for {column}: {cramers_value:.2f}")
```

```python
# Loop through each categorical column
for column in categorical_columns:
    # Create a stacked bar chart
    contingency_table = pd.crosstab(df_cleaned[column], df_cleaned['loan_status'], normalize='index')
    contingency_table.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='coolwarm')

    # Add labels and title
    plt.title(f'Loan Status by {column.replace("_", " ").title()}', fontsize=16, fontweight='bold')
    plt.xlabel(column.replace('_', ' ').title(), fontsize=14)
    plt.ylabel('Proportion', fontsize=14)
    plt.legend(title='Loan Status', loc='upper right', labels=['Not Approved (0)', 'Approved (1)'])
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
# List of categorical variables for analysis
group_columns = ['person_gender', 'person_home_ownership', 'loan_intent', 'person_education', 'previous_loan_defaults_on_file']

# Loop through each categorical variable
for column in group_columns:
    # Calculate loan approval rates
    approval_rates = df_cleaned.groupby(column)['loan_status'].mean() * 100  # Convert to percentage

    # Display results
    print(f"Loan Approval Rates by {column.title().replace('_', ' ')}:")
    print(approval_rates.round(2))
    print('-' * 50)
```

```python
import pandas as pd
import numpy as np


# Capping outliers for selected columns
def cap_outliers(column, lower_percentile=0.01, upper_percentile=0.99):
    lower_bound = column.quantile(lower_percentile)
    upper_bound = column.quantile(upper_percentile)
    column = np.where(column < lower_bound, lower_bound, column)
    column = np.where(column > upper_bound, upper_bound, column)
    return column

columns_with_outliers = ['person_age', 'person_income', 'loan_amnt', 'loan_int_rate',
                         'cb_person_cred_hist_length', 'credit_score']
for col in columns_with_outliers:
    df[col] = cap_outliers(df[col])

# Handling high correlation
df.drop(columns=['cb_person_cred_hist_length'], inplace=True)  # Dropping one correlated variable

# Handling rare categories
df['person_home_ownership'] = df['person_home_ownership'].replace({'OTHER': 'RENT'})  # Group 'OTHER' into 'RENT'

# Convert Boolean columns to numeric
df['person_gender'] = df['person_gender'].map({'female': 0, 'male': 1})
df['previous_loan_defaults_on_file'] = df['previous_loan_defaults_on_file'].map({'no': 0, 'yes': 1})

# Verify changes
print(df.head())
```
